```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Transpose to get adoption rates as a single column
adoption_rates = df.T[0]

# Create time steps as index
steps = adoption_rates.index

# Plot adoption rate over time
plt.plot(steps, adoption_rates)
plt.xlabel("Time step")
plt.ylabel("Adoption rate")
plt.title("Diffusion dynamics")
plt.show()
```
